# Loops
Loops are useful when you want to repeat the same thing many times on different values.

### Definitions

**For loop:** Use when you want to execute a block of code **for** each element in a string, list, dictionary, etc. (known # of repetitions)

**While loop:** Use when you want to execute a block of code **while** a certain continuation criteria is met (unknown # of repetitions)

**Iterate:** Perform repeatedly

## While loops

General form of a while loop:

`while condition is true:
    do something`

```python
Image('http://3.bp.blogspot.com/-i1W59d7B8R4/T3izXShRgHI/AAAAAAAAAFI/OcmcydxcZV0/s1600/while.jpg')
```

```python
# How many times in a row is a random number chosen between 1 and 100 less than 50?

# Generate a random number and print it out
random_number = random.randint(1,101)
print(random_number)

# Start the count at zero
count = 0

# When the random number is less than 50, increase the count and generate a new random number
while random_number < 50:
    count += 1
    random_number = random.randint(1,101)
    print(random_number)
    
print('Count:',count)
```

```python
# Make it into a function
# Function to return the number of times in a row a random number is less than a given number
def random_less_than(number,minimum,maximum):
    # Generate a random number and print it out
    random_number = random.randint(minimum,maximum)
    
    # Start the count at zero
    count = 0

    # When the random number is less than 50, increase the count and generate a new random number
    while random_number < number:
        count += 1
        random_number = random.randint(minimum,maximum)

    return count

# Call function
count_less_50 = random_less_than(50,1,101)
print('Count:',count_less_50)
```

```python
# What happens if we do this 1000 times? How often do you expect the count to be 0? How about 1?

# Make a variable to store the counts
counts = []

# Ask the user how many times we should run the function
#n_times = int(input("Number of iterations? "))
n_times = 1000
# Run the function that many times
for i in range(n_times):
    count = random_less_than(50,1,101)
    # Append the result to counts each time
    counts.append(count)
    
# Plot a histogram of the counts
plt.hist(counts,bins=len(set(counts)),align='left')
plt.xlabel('Counts')
plt.ylabel('Frequency')
```
